```python
import pandas as pd
import numpy as np

np.random.seed(1)
no_sample=1000
no_class_0_ratio=0.9
no_class_0=int(no_sample *no_class_0_ratio)
no_class_1=100
```

```python
no_class_0,no_class_1
```

```python
class_0={'feature1':np.random.normal(0,1,no_class_0),'feature2':np.random.normal(0,1,no_class_0),'target':[0]*no_class_0}
```

```python
class_0=pd.DataFrame(class_0)
```

```python
class_0
```

```python
class_1={'feature1':np.random.normal(3,1,no_class_1),'feature2':np.random.normal(3,1,no_class_1),'target':[1]*no_class_1}
```

```python
class_1=pd.DataFrame(class_1)
```

```python
class_1
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```
